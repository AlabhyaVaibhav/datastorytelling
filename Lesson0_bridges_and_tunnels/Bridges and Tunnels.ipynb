The MTA makes [toll plaza collection data](http://mta.info/developers/data/bandt/trafficdata.html) available on a weekly basis. I downloaded this data and reduced it to a file containing only the number of cash and electronic payments by day for the Verrazano Bridge (one of New York's prettiest!).

The data is in CSV format with each row having: date, # cash payments, # electronic payments. You can look at the [raw data file on github](https://github.com/hmason/datastorytelling/blob/master/Lesson0_bridges_and_tunnels/verrazano.csv). 

We can start this analysis by using the [csv2rec](http://matplotlib.org/api/mlab_api.html#matplotlib.mlab.csv2rec) function to load the data from the file into a record that we can manipulate.

```python
verrazano_data = mlab.csv2rec("verrazano2.csv")
```

Our data has been magically transformed into an object where each row is a date, followed by the number of cash transactions and the number of electronic transactions.

What questions might we ask now?

How about the basics? How many people, on average, pay cash at the Verrazano bridge each day?

```python
cash_payments = [int(f[1]) for f in verrazano_data] # create a list that is a subset of our data including *only* cash payments
numpy.mean(cash_payments) # calculate the mean
```

That\'s the average, but what about the minimum and maximum values?

```python
min(cash_payments)
```

```python
max(cash_payments)
```

...and how many people pay electronically?

```python
etc_payments = [int(f[2]) for f in verrazano_data]
numpy.mean(etc_payments)
```
