The MTA makes [toll plaza collection data](http://mta.info/developers/data/bandt/trafficdata.html) available on a weekly basis. I downloaded this data and reduced it to a file containing only the number of cash and electronic payments by day for the Verrazano Bridge (one of New York's prettiest!).

We can start by using the csv2rec function to load the data from the file into a record that we can manipulate.

```python
verrazano_data = mlab.csv2rec("verrazano.csv")
```

Now let's take a look at a random row from the dataset (and you can re-run that cell a few times to see different samples).

What questions might we ask now?

How many people, on average, pay cash at the Verrazano bridge each day?

```python
cash_payments = [int(f[1]) for f in verrazano_data] # create a list that is a subset of our data including *only* cash payments
numpy.mean(cash_payments)
```

```python
That's the average, but what about the minimum and maximum values?
```

...and how many people pay electronically?

```python
etc_payments = [int(f[2]) for f in verrazano_data]
numpy.mean(etc_payments)
```
